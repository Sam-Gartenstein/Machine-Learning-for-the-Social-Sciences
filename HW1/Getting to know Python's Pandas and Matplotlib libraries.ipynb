```python
#Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. Find the url for the mtcars dataset from the following website:

https://vincentarelbundock.github.io/Rdatasets/datasets.html

Read through the "DOC" file to understand the variables in the dataset, then use the following url to import the data using pandas read_csv function.  


```python
#Creating dataframe for car data, which I will refer to as car dataframe
car_df = pd.read_csv("https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv")
```

2.  Display the first five rows of the data.

```python
#Displaying the first five rows of the car dataframe
car_df.head()
```

3. Calculate the average of the mpg column for all cars within each category of the cyl column.





```python
#Grouping the car dataframe by cyl and mpg and calculating  the mpg average for all cars within each category of cyl
car_df.groupby('cyl')[['mpg']].mean()
```

4.  Create a histogram using the mpg column

```python
#Creating histogram for mpg in car dataframe
car_df.hist(column='mpg', bins=5, grid=True)

#Adding a title, x-label, and y-label to the histogram
plt.title("Histogram of Miles Per Gallon (MPG)")
plt.xlabel("mpg")
plt.ylabel("Frequency")
```

5. Choose two variables in the data and create a scatterplot.



```python
#scatter plot of weight and mpg, where weight is in the x-axis and mpg is in the y-axis
car_df.plot.scatter(x='wt', y='mpg')
```
